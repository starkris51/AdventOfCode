```polyglot-notebook
#load "../utils.fsx"
open Utils
let input = GetData 6 |> GetLines
let opLine = input |> Array.last
let numberLines = input |> Array.take (input.Length - 1)

let numberList = numberLines |> Array.map (fun line -> line.Split([|' '|], System.StringSplitOptions.RemoveEmptyEntries) |> Array.map int64)

let operationList = opLine.Split([|' '|], System.StringSplitOptions.RemoveEmptyEntries)
numberList
```

```polyglot-notebook
let rowLength = numberList[0].Length

let results =
    [0 .. rowLength - 1]
    |> List.map (fun i -> 
        let operation = operationList[i]    
        let numbers = numberList            
                      |> Array.map (fun row -> row[i])
        match operation with
        | "+" -> numbers |> Array.sum
        | "*" -> numbers |> Array.reduce (*)
        | _ -> failwith "Unknown operation"
    )
results |> List.sum
```
