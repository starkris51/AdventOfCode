```polyglot-notebook
#load "../utils.fsx"
open Utils
let input: string[] = GetData 2 |> Split(",")
```

Part 1

```polyglot-notebook
let mutable invalidIds: int64[] = [||]

input
|> Array.iter (fun idRanges -> 
    let firstId = idRanges.Split('-').[0] |> int64
    let lastId = idRanges.Split('-').[1] |> int64
    
    for id in firstId .. lastId do
        let s = string id
        let isInValid = 
            if s.StartsWith("0") then false
            elif s.Length % 2 <> 0 then false
            else 
                let half = s.Length / 2
                let firstHalf = s.[0 .. half - 1]
                let secondHalf = s.[half .. s.Length - 1]

                firstHalf = secondHalf
        if isInValid then
           invalidIds <- Array.append invalidIds [| id |]
        ()
    ()
)

invalidIds |> Array.sum |> printfn "%d"
```

```polyglot-notebook
let mutable invalidIds: int64[] = [||]

input
|> Array.iter (fun idRanges -> 
    let firstId = idRanges.Split('-').[0] |> int64
    let lastId = idRanges.Split('-').[1] |> int64
    
    for id in firstId .. lastId do
        let s = string id
        let isInValid = 
            if s.StartsWith("0") then false
            else 
                let mutable allSame = false
                for division in 2 .. s.Length do
                    if s.Length % division = 0 then
                        let partLen = s.Length / division
                        // Build the divided strings array
                        let dividedStrings =
                            [| for i in 0 .. division - 1 ->
                                   s.Substring(i * partLen, partLen) |]
                        // Check if all divided strings are the same
                        let firstDividedString = dividedStrings.[0]
                        if dividedStrings |> Array.forall (fun ds -> ds = firstDividedString) then
                            allSame <- true
                allSame


        if isInValid then
           invalidIds <- Array.append invalidIds [| id |]
        ()
    ()
)

invalidIds |> Array.sum |> printfn "%d"
```
