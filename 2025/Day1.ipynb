Load some stuff

```polyglot-notebook
#load "../utils.fsx"
open Utils
let input: string[] = GetData 1 |> GetLines
input
```

Part 1

```polyglot-notebook
let mutable dial = 50
let mutable numberOfZeros = 0

for rotation in input do
    let direction = rotation.[0]
    let amount = rotation.[1..] |> int
    
    if direction = 'L' then
        dial <- (dial - amount) % 100
    else
        dial <- (dial + amount) % 100

    if dial = 0 then
        numberOfZeros <- numberOfZeros + 1

numberOfZeros
```

Part 2

```polyglot-notebook
let mutable dial2 = 50
let mutable numberOfZeros2 = 0

for rotation in input do
    let direction2 = rotation.[0]
    let amount2 = rotation.[1..] |> int
    
    if direction2 = 'L' then
        for i in 1 .. amount2 do
            dial2 <- (dial2 - 1) % 100
            if dial2 = 0 then
                numberOfZeros2 <- numberOfZeros2 + 1
    else
        for i in 1 .. amount2 do
            dial2 <- (dial2 + 1) % 100
            if dial2 = 0 then
                numberOfZeros2 <- numberOfZeros2 + 1

numberOfZeros2
```
