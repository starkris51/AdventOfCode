```polyglot-notebook
#load "../utils.fsx"
open Utils
// convert to 2d array
let mutable input = GetData 7 |> Split("\n") |> Array.map (fun x -> x.ToCharArray() |> Array.map string)
input
```

```polyglot-notebook
let mutable splitsCount = 0
// Find S in the 2d array
let startingIndex : int = input[0] |> Array.findIndex (fun x -> x = "S")

let mutable beams = [|startingIndex|]

for rowIndex in 1 .. input.Length - 1 do
    let row = input[rowIndex]
    for colIndex, col in row |> Array.indexed do
        if Array.contains colIndex beams || input[rowIndex - 1][colIndex] = "S"  then
            input[rowIndex][colIndex] <- "|"
        if col = "^" && input[rowIndex - 1][colIndex] = "|" && Array.contains colIndex beams then
            splitsCount <- splitsCount + 1
            beams <- Array.filter (fun x -> x <> colIndex) beams
            beams <- Array.append beams [|colIndex - 1; colIndex + 1|]

splitsCount
```
