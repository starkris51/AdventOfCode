```polyglot-notebook
#load "../utils.fsx"
open Utils
let input = GetData 5 |> GetLines
input

let idRanges = input |> Array.filter(fun line -> line.Contains("-"))

let ids = input |> Array.filter(fun line -> not (line.Contains("-")))
```

```polyglot-notebook
let isInRange (range: string, value: int64) =
    let parts = range.Split("-") |> Array.map int64
    let first, last = parts.[0], parts.[1]
    value > first && value < last

let checkFresh (id: string): bool =
    match System.Int64.TryParse id with
    | true, idValue ->
        idRanges
        |> Array.exists (fun range -> isInRange (range, idValue))
    | false, _ ->
        false

let freshIds = ids |> Array.filter checkFresh
freshIds.Length
```

```polyglot-notebook
let ranges : (int64 * int64)[] =
    idRanges
    |> Array.map (fun range ->
        let parts = range.Split("-") |> Array.map int64
        let first, last = parts.[0], parts.[1]
        first, last
    )

// this part 2 too hard

```
