```polyglot-notebook
#load "../utils.fsx"
open Utils
// convert to array 2d
let input = GetData 4
let to2DArray (lines: string) =
    lines.Split("\n")
    |> Array.map (fun line -> line.ToCharArray() |> Array.map string)
let parsedInput: string[][] = to2DArray input

let height = parsedInput.Length
let width = parsedInput.[0].Length

let inBounds (x, y) =
    x >= 0 && x < width && y >= 0 && y < height

let getNeighbors (x,y) =
    [ 
        (x-1,y)
        (x-1,y-1)
        (x-1,y+1)
        (x+1,y)
        (x+1,y-1)
        (x+1,y+1)
        (x,y-1)
        (x,y+1)
    ]
    |> List.filter inBounds
parsedInput
```

```polyglot-notebook
let countPaperRolls (parsed: string[][]): int =
    let mutable number = 0
    for y in 0 .. height - 1 do
        for x in 0 .. width - 1 do
            if parsed.[y].[x] = "@" then
                let neighbors = getNeighbors (x,y)
                let mutable amountOfNeighborPaperRolls = 0
                neighbors |> List.iter (fun (nx, ny) -> 
                if parsed.[ny].[nx] = "@" then
                    amountOfNeighborPaperRolls <- amountOfNeighborPaperRolls + 1
                )
                
                if amountOfNeighborPaperRolls < 4 then
                    number <- number + 1


    number

countPaperRolls parsedInput
```

```polyglot-notebook
let mutable current = parsedInput
let mutable totalRemoved = 0

let removePaperRolls (input: string[][]): int * string[][] =
        
    let copy =
        input
        |> Array.map (fun row -> Array.copy row)

    let mutable removedThisRound = 0

    for y in 0 .. height - 1 do
        for x in 0 .. width - 1 do
            if input.[y].[x] = "@" then
                let neighbors = getNeighbors (x,y)
                let mutable neighborCount = 0
                neighbors |> List.iter (fun (nx, ny) -> 
                if copy.[ny].[nx] = "@" then
                    neighborCount <- neighborCount + 1
                )
                if neighborCount < 4 then
                    copy.[y].[x] <- "x"
                    removedThisRound <- removedThisRound + 1
    removedThisRound, copy

let mutable keepGoing = true

while keepGoing do
    let removed, newGrid = removePaperRolls current
    if removed = 0 then
        keepGoing <- false
    else
        totalRemoved <- totalRemoved + removed
        current <- newGrid

totalRemoved
```
