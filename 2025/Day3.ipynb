```polyglot-notebook
#load "../utils.fsx"
open Utils
let input: string[] = GetData 3 |> GetLines
input
```

Part 1

```polyglot-notebook
let mutable joltages = [||]

input
|> Array.iter (fun (bank: string) -> 
    let mutable firstDigit = 0
    let mutable secondDigit = 0
    let mutable maxJoltageForBank = -1

    for i in 0 .. bank.Length - 2 do
        let number = int bank.[i] - int '0'

        for j in i+1 .. bank.Length - 1 do
            let nextNumber = int bank.[j] - int '0'
            if int (string number + string nextNumber) > maxJoltageForBank then
                maxJoltageForBank <- int (string number + string nextNumber)
                ()
    
    let fullJoltage = string maxJoltageForBank
    joltages <- Array.append joltages [|int fullJoltage|]
    ()
)

joltages |> Array.sum |> string
```

Part 2

```polyglot-notebook
let mutable joltages = [||]

input
|> Array.iter (fun (bank: string) -> 
    let mutable startIndex = 0
    let mutable chosen = ""
    let mutable remainingDigits = 12

    while remainingDigits > 0 do
        let mutable bestDigit = -1
        let mutable bestIndex = startIndex
        let maxIndex = bank.Length - remainingDigits

        for i in startIndex .. maxIndex do
            let candidate = int bank.[i] - int '0'
            if candidate > bestDigit then
                bestDigit <- candidate
                bestIndex <- i
                ()

        chosen <- chosen + string bestDigit
        startIndex <- bestIndex + 1

        remainingDigits <- remainingDigits - 1
        ()
    
    joltages <- Array.append joltages [|int64 chosen|]
    ()
)

joltages |> Array.sum |> string
```
