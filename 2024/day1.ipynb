```polyglot-notebook
let parseList (input: string): int[] * int[] = 
    input
    |> GetLines
    |> Array.map (fun x -> Split " " x |> Array.map int)
    |> Array.transpose
    |> fun x -> x[0], x[1]
```

Part 1

```polyglot-notebook
let part1 (a: int[]) (b: int[]): int =
    let list1 = Array.sort a
    let list2 = Array.sort b
    Array.zip list1 list2
    |> Array.sumBy (fun (x, y) -> abs (x - y))

let a,b: int[] * int[] = parseList input

let result: int = part1 a b
result
```

Part 2

```polyglot-notebook
let part2 (a: int[]) (b: int[]) =
    let mutable sum = 0
    for left in a do
        let mutable amount = 0
        for right in b do
            if left = right then
                amount <- amount + 1
        sum <- sum + (left * amount)
    sum

let result2: int = part2 a b
result2
```
